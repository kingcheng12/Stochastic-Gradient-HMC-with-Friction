```python
import scipy.stats as sta
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from numba import jit,vectorize
from ipyparallel import Client
import random
import numpy
```

```python
from sghmc.module import SGHMC, SGHMC_parallel
```

```python
mu = np.array([-5, 5]).reshape(2,1) # true value for mu
n = 200 # number of observations
y = np.r_[np.random.normal(mu[0], 1, n),np.random.normal(mu[1], 1, n)]
```

```python
def gradU2(scale,theta,batch):
    def g_l_pdf(y,mu):
        e1=numpy.exp(-0.5*(y-mu[0])**2)
        e2=numpy.exp(-0.5*(y-mu[1])**2)

        v = numpy.c_[e1*(y-mu[0]),e2*(y-mu[1])]
        return numpy.sum(v/(e1+e2),axis=0)

    def g_l_pr(mu):
        return -(numpy.sum(mu)/100)
    return - scale*g_l_pdf(batch,theta)-g_l_pr(theta)
```

```python
theta1 = SGHMC(theta0=np.array([1,1]), X = y[:,None], gradU = gradU2, eps=0.01, sample_size=1000, B=0, C=np.eye(2), batch_size=100,burnin=500)
```

```python
sns.kdeplot(theta1[:,0],theta1[:,1])  
plt.title('density plot of mixture normal')
plt.savefig('mixture_norm.png')
```

```python
theta2 = SGHMC_parallel(theta0=np.array([1,1]), X = y[:,None], gradU = gradU2, eps=0.01, sample_size=1000, B=0, C=np.eye(2), batch_size=100,burnin=500)
```

```python
sns.kdeplot(theta2[:,0],theta2[:,1])  
plt.title('density plot of mixture normal with parallel')
plt.savefig('mixture_norm_parallel.png')
```
